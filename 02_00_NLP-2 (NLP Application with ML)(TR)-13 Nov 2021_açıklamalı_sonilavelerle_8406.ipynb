```python
stop_words = stopwords.words('english')
```

```python
def cleaning(data):
    
    #1. Tokenize
    text_tokens = word_tokenize(data.replace("'", "").lower()) 
        # ayraçları kaldırıyorum ki (can't gibi) olumsuzluk içeren kelimeler içinde ayraç içerdiği için stopwords lerde silinmesin.
        # Yorumun olumsuz haliyle de olsa kalması için ilave bir yöntem olarak kullanıldı.
        
    #2. Remove Puncs
    tokens_without_punc = [w for w in text_tokens if w.isalpha()]  
        # punctuation ve no lardan kurtuluyoruz.
    
    #3. Removing Stopwords
    tokens_without_sw = [t for t in tokens_without_punc if t not in stop_words]  # stopword lerden temizleme
        # tokenleri teker teker al stopwords ler içinde yoksa olduğu gibi yaz, yani stopwords ise yazma.
        # storwards olmayanları liste içine yazmaya devam et   
    
    #4. lemma
    text_cleaned = [WordNetLemmatizer().lemmatize(t) for t in tokens_without_sw]
        # stopwardslardan temizlenmiş tokenleri alıyoruz ve lemma uyguluyoryuz
        
    #joining
    return " ".join(text_cleaned)
        # tamamen temizlenmiş ve köklerine inilmiş kelimeleri alıp join yapıyoruz.
```

```python
# def açıklaması:
# - iki tane def te aynı, aşağıdaki bir önceki derste gösterilendir.
# - yukarıdaki def fonksiyonunda ayraçları kaldırma işlemi gerçekleştirdik. upgrade ettik.
# - 1. ayraçlar kaldırıldı ve küçük harfe dönüştürüldü.
# - 2. noktalama işaretler ve numaralardan arındırılma işlemi yapıldı
# - 3. token without punc içindeki tüm tokenleri al ve içinden stop words olmayanlar ile yola devam et.
# - 4. stop word lerden arındırılmış tokenlerin lemma yani köklerine gittik.
# - 5. text imizi birleştirdik.
```

```python
" I don't like".replace("'", "")   # Yorumun olumsuz haliyle de olsa kalması için ilave bir yöntem olarak kullanıldı.
```

## Model Comparisons - Vectorization

```python
from sklearn.metrics import plot_confusion_matrix,classification_report, f1_score, recall_score
```

```python
def eval(model, X_train, X_test):
    y_pred = model.predict(X_test)
    y_pred_train = model.predict(X_train)
    print("Test_Set")
    print(classification_report(y_test,y_pred))
    print("Train_Set")
    print(classification_report(y_train,y_pred_train))
    plot_confusion_matrix(model,X_test, y_test)
```

## Naive Bayes

```python
# A olayı gerçekleştiğinde B olayının gerçekleşme olasılığı nedir? (P(A|B))
# B olayı gerçekleştiğinde A olayının gerçekleşme olasılığı nedir? (P(B|A))
# Datasetime nasıl uygulayabilirim?
# P(dont | negative) --> dont yardımcı fiili geçtiğinde bunun negatif olma ihtimali nedir
# P(dont enjoy|positive) --> dont ve enjoy tokenleri birlikte geçtiğinde positive olma ihtimali nedir
```

## TF-IDF

```python
# scale işleminde yaptığımız gibi, leakage önlemek için train --> fit_transform, test --> transform


# fit_transform dediğimiz için unique olan tüm kelimeleri sayıyor, hangi satırlarda geçtiğini buluyor ve array e dönüştürüyor
# test te geçen bir kelime train de geçmiyorsa model o kelimeyi yok sayar, görmez.
# model ne kadar fazla kelime içeriyorsa o kadar bu handikapın önüne geçiyor.

# TF
# FIT train setindeki unique tokenleri tespit ediyor ve hangi satırlarda geçtiğini buluyor, 
# transform bu tokenleri ve geçtiği satırları sayıyor
# FİT train setine göre yapıldığı için, TEST te TRAİNe göre transform olacak.
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# eğitim mantığı diğerleriyle tamamen aynı

tf_idf_vectorizer = TfidfVectorizer()
X_train_tf_idf = tf_idf_vectorizer.fit_transform(X_train) # FIT train setindeki unique tokenleri tespit ediyor ve hangi satırlarda geçtiğini buluyor
X_test_tf_idf = tf_idf_vectorizer.transform(X_test)       # TRANSFORM bu tokenlerin sayısını alıyor ve bu tokenlerle TF-IDF yapıyor

# tf_idf: fit deyince X_traindeki tüm unique tokenleri tespit ediyor ve bütün satırlarda (dokümanlarda) o tokenin geçme sıklığını da sayıyor.
# count vectorizer da sadece unique tokenleri tespit ediyordu burda dokümanlardaki tekrar sayısı da işin içinde..

# transform deyince bunları sayıyor ve fonksiyonu uygulayıp tf-idf katsayılarını buluyor.
```

```python
# TF = KELİMEM / TOPLAM KELİME SAYISI
# DF = CORPUS TAKİ TOPLAM DOKÜMAN SAYISI / KELİMEMİN GEÇTİĞİ DOKÜMAN SAYISI
```

```python
# TF(t) = (Number of times term t appears in a document) / (Total number of terms in the document)

# IDF(t) = log(Total number of documents  in a corpus/ Number of documents with term t in it)
```

```python
# Example: 
# Consider a document containing 100 words wherein the word cow appears 3 times
# The TF for cow is then (3 / 100) = 0.03 
# Now, let's assume the corpus has 10 million documents and the word cow appears in one thousand of these
# Then, the IDF is calculated as log(10,000,000 / 1,000) = 4
# Thus, the Tf-idf weight is the product of these quantities: 0.03 * 4 = 0.12
```

```python
Öncelikle elimizde 4 tane döküman olsun.

doc1 = "a a b c"
doc2 = "a c c c d e f"
doc3 = "a c d d d"
doc4 = "a d f"

d anahtar kelimesinin 3.dökümanı/veya satirini ne denli temsil ettiğini bulalım yani tf-idf değerini bulalım.

key = "d"

TF:
TF = doc3.count(key) / len(doc3)  # 3 / 5 == 0.6


Kelimenin kaç dökümanda geçtiğini(count) hesaplayalım.
IDF:

all_documents = [doc1, doc2, doc3, doc4] # tüm dökümanları birleştir.
count = 0
for doc in all_documents:
    for word in doc:
        if word == key:
            count += 1
            break
# count = 3

Toplam döküman/satir sayısını ve bulduğumuz count değeri logaritmaya sokarak IDF değerini hesaplayalım.

total_doc_number = len(all_documents)
DF = total_doc_number / count
from math import log10
IDF = log10(DF) # 0.12493873660829993
Ve artık bulduğumuz sonuçları çarparak TF-IDF değerini görelim.

TF * IDF  # 0.07
```
